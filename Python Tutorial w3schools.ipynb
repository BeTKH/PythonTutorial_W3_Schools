### Data Types in Python

**Variables**  can store data of *different types* , and different types can do different things.

```python
#by Blu
#Data Types in Python

#++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

print("1. Str type", '\n','======================')
#Str


x = 'Hey'
print(type(x),x,'\n')

#++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
print("2. Numeric Types", '\n','======================')
#int, float and complex

y,z = 1, 2.3
print(type(y),y)
print(type(z),z)
a = 2 + 5j  #note 'j' NOT 'i'
print(type(a), a,'\n')

#++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
print("3. Sequence Types", '\n','======================')
#sequence type

mylist = [1,2,"3"]
mytup = (1,2,"3")
myrange = range(5)

print(type(mylist), mylist)
print(type(mytup), mytup)
print(type(myrange), myrange, '\n')

#++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
print("4. Mapping Type", '\n','======================')
#Dictionary

mydict = {'name': 'Blu', 'age': '28', 'job':'student'}
print(type(mydict),mydict, '\n')

#++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
print("5. Set Type", '\n','======================')
#Set, Frozen set

myset = {1,2, 3}
myfrozenset = frozenset({1, 2, 'apple', 'book'}) #frozenset([iterable])


print(type(myset), myset)
print(type(myfrozenset),myfrozenset,'\n')

#++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
print("6. Boolean type", '\n','======================')
#Boolean Type

b1 = True
#b2 = true   'This is undefined type'
b3 = 'true'  # this is simply str type
print(type(b1), b1)
#print(type(b2))
print(type(b3),b3, '\n')

#++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

print("7. Binary type", '\n','======================')

#Bytes
by1 = bytes(10)
print(type(by1),by1)
by2 = b"10"
print(type(by2),by2)

#Bytearray
bya = bytearray(10)

print(type(bya),bya)


#Memoryview
memview = memoryview(bytes(10))
print(type(memview),memview, '\n')

#++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
```

#### Identity Operations

    key words: 
           
           is
           is not

```python
x = 2
y = 2
z = x is y
print(z)
```

```python
x = [1,2,3]
y = x
z = y is x
print(z)  #True means any change made to y will also change x.
```

```python
x = [1,2,3]
y = x.copy()
z = y is x
print(z) #False means any change made to y will NOT change x.
```

### 9.	Python Lists

#### Python Collections (Arrays)

There are four collection data types in the Python programming language:

    •	[]List - is a collection which is ordered and changeable. 
             - Allows duplicate members.
             - name = [member, member, ..., menber]
               
    •	()Tuple - is a collection which is ordered and unchangeable. 
              - Allows duplicate members.
              - name = (member, member, member,....member)
                
    •	{}Set - is a collection which is unordered and unindexed. 
            - No duplicate members.
            -setname = (member, member, member,....member)
              
    •	{}Dictionary - is a collection which is unordered, changeable and indexed. 
                   - No duplicate   members.
                   -name = {'id': 'member', 'id2': 'member', 'id3':'member',....'idn': 'member'}

When choosing a collection type, it is useful to understand the properties of that type. Choosing the right type for a particular data set could mean retention of meaning, and, it could mean an increase in efficiency or security.

```python
mylist = ["apple", "banana", "cherry"]
mylist2 = mylist[1:]
print(mylist2)
mylist3 = mylist2[1:]
print(mylist3)
#together
print(mylist3 == mylist[1:][1:])
print(mylist3 is mylist[1:][1:])
```

```python
print(mylist[-1][2:5])
print(mylist[1:][1:])
print(mylist[1:2])     #mylist[included:NotIncluded]
print(mylist[-3:-1])   #mylist[included:NotIncluded]
```
